### 1.3 Standardize Data
- We need to convert the years into datetime format.

````python
## copy the instance for safe keeping

df_sd = df.copy()
````

#### We need to convert the years into datetime format.


````python
## first we rename the Year Started and Year Ending (Estimate)
df_sd.rename(columns={'Year Started': 'Start Date', 'Year Ending (Estimate)': 'Estimated End Date'}, inplace=True)

df_sd.rename(columns={'Total Value of Project ': 'Total Value'}, inplace=True)
## Convert start date to integer, handling missing values using the mean
# Calculate the mean of 'Start Date' (excluding NaN values)
mean_start_date = df_sd['Start Date'].dropna().mean()

# # Fill NaN values with the calculated mean and convert to integer
df_sd['Start Date'] = df_sd['Start Date'].fillna(mean_start_date).astype(int)

df_sd.head()
````

####  convert all foreign currency to USD and MZN

````python
# check the unique currencies
df_sd['Currency'].unique()
````

````python
import requests

API_KEY = "741475462d95e6d097c23e7e"  # Get a free key from exchangerate-api.com
BASE_URL = f"https://v6.exchangerate-api.com/v6/{API_KEY}/latest/"

def convert_to_usd(row):
    try:
        currency = str(row['Currency']).strip().upper()
        amount = float(row['Total Value'])

        if currency == 'USD':
            return amount

        response = requests.get(BASE_URL + currency)
        data = response.json()

        if response.status_code == 200 and 'conversion_rates' in data:
            rate = data['conversion_rates'].get('USD')
            if rate:
                return amount * rate

        print(f"Error fetching rate for {currency}")
        return None

    except Exception as e:
        print(f"Error converting {row['Currency']} {row['Total Value']}: {e}")
        return None

df_sd['Total Value (USD)'] = df_sd.apply(convert_to_usd, axis=1)
````

### 2. Treemap displaying funding partners proportional to total project value

````python
!pip install plotly
````

````python
# Calculate the total value of projects for each funding partner
partner_value_counts = df_clean.groupby('Funding Partner')['Total Value (USD)'].sum()

# round it down to one digit
partner_value_counts = partner_value_counts.round(1)

# Create a DataFrame for the treemap
treemap_data = pd.DataFrame({'Funding Partner': partner_value_counts.index, 'Total Value': partner_value_counts.values})
treemap_data.sort_values('Total Value', ascending=False, inplace=True)
treemap_data.reset_index(drop=True, inplace=True)
treemap_data
````

````python
import plotly.express as px

# Create the treemap
fig = px.treemap(
    treemap_data,
    path=['Funding Partner'],  # Hierarchical levels (just one in this case)
    values='Total Value',
    title='Total Funding Distribution by Partner',
    color='Total Value',
    color_continuous_scale='Blues'
)

# Show the treemap
fig.show()
````

### 4. Line chart showing total project value over time (by start year)



````python
### Line chart showing total project value over time (by start year)
import matplotlib.ticker as ticker

# Convert 'Start Date' to datetime and extract year
df_clean['Start Date'] = pd.to_datetime(df_clean['Start Date'], format='%Y')
df_clean['Start Year'] = df_clean['Start Date'].dt.year

# Group by Start Year and sum Total Value (USD)
yearly_totals = df_clean.groupby('Start Year')['Total Value (USD)'].sum()

# Create area chart
plt.figure(figsize=(10, 6))
plt.fill_between(yearly_totals.index, yearly_totals.values, color='skyblue', alpha=0.5)  # Shaded area

# Formatting the Y-axis to show values in billions (1e9)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'${x/1e9:.1f}B'))

plt.title('Total Project Value Over Time (by Start Year)')
plt.xlabel('Start Year')
plt.ylabel('Total Value (USD)')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
````

*Observation: The highest project funding occurred in 2024, indicating a significant increase in investment compared to previous years. This suggests a growing emphasis on project development, possibly due to increased budget allocations, economic growth, or strategic priorities. The sharp rise in funding may also reflect new policies, investor confidence, or large-scale projects initiated during this period.*

### 8. HeatMap representaton of Collaboration Patterns Between Funding Partners and Implementing members

````python
# Visualization:
# Create a collaboration matrix
collaboration_matrix = df_clean.groupby(['Funding Partner', 'Implementing Partner Name ']).size().head(10).unstack(fill_value=0)

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(collaboration_matrix, annot=True, cmap="YlGnBu", fmt="d")
plt.title('Collaboration Patterns: Funding Partners vs. Implementing Partners')
plt.xlabel('Implementing Partner')
plt.ylabel('Funding Partner')
plt.show()
````

The heatmap illustrates the collaboration patterns between funding partners and implementing partners. Notably:

- **Bill and Melinda Gates Foundation** collaborates with multiple implementing partners, including CHAI, Fundação Manhiça, Instituto Nacional de Saúde, The Global Fund, and WHO. This suggests a broad investment strategy focusing on diverse projects.
- **Canada** exclusively collaborates with Aga Khan Foundation, Ipas, Oxfam, and Pathfinder International, indicating a selective partnership approach.
- **Arab Bank for Economic Development in Africa** appears to have a single collaboration, limiting its engagement with implementing partners compared to other funders.
This visualization highlights distinct funding strategies, where some partners distribute funds widely while others maintain targeted partnerships.

### 10. How did project funding change during and after the COVID-19 pandemic (2020-2023)?

````python
df_clean.head()
````

````python
# How did project funding change during and after the COVID-19 pandemic (2020-2023)?
# Convert 'Start Date' to datetime and extract the year
df_clean['Start Date'] = pd.to_datetime(df_clean['Start Date'], format='%Y')
df_clean['Start Year'] = df_clean['Start Date'].dt.year

# Filter data for the period 2020-2023
pandemic_period_df = df_clean[(df_clean['Start Year'] >= 2020) & (df_clean['Start Year'] <= 2023)]

# Group by year and sum the total funding
funding_by_year = pandemic_period_df.groupby('Start Year')['Total Value (USD)'].sum()

# Calculate percentage change from the previous year
percentage_change = funding_by_year.pct_change() * 100
percentage_change = percentage_change.round(2)

# Create a DataFrame for the percentage change
percentage_change.columns = ['Year', 'Percentage Change']

# create dataframe
percentage_change = percentage_change.reset_index()
percentage_change

# Create an area chart
plt.figure(figsize=(10, 6))
plt.fill_between(percentage_change['Start Year'], percentage_change['Total Value (USD)'], color='skyblue', alpha=0.5)

# Formatting the chart
# plt.gca().yaxis.set_major_formatter(ticker.PercentFormatter())
plt.title('Percentage Change in Project Funding During and After COVID-19 Pandemic (2020-2023)')
plt.xlabel('Year')
plt.ylabel('Percentage Change from Previous Year')
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.show()
````
